# ***Novo dataframe (óbitos)***

```python
result = df2.groupby(['NU_ANO','SG_UF_NOT'])['EVOLUCAO'].sum().reset_index()
anos = df['NU_ANO'].unique()
estados = df['SG_UF_NOT'].unique()
for ano in anos:
  for estado in estados:
    if not ((result['NU_ANO'] == ano) & (result['SG_UF_NOT'] == estado)).any():
      result = result.append({'NU_ANO': ano, 'SG_UF_NOT': estado, 'EVOLUCAO': 0}, ignore_index=True)





result = result.rename(columns={'NU_ANO': 'Ano'})
result = result.rename(columns={'SG_UF_NOT': 'Sigla'})
result = result.rename(columns={'EVOLUCAO': 'Óbitos'})
result


siglas_estados = {
    'RO': 'Rondônia',
    'AC': 'Acre',
    'AM': 'Amazonas',
    'RR': 'Roraima',
    'PA': 'Pará',
    'AP': 'Amapá',
    'TO': 'Tocantins',
    'MA': 'Maranhão',
    'PI': 'Piauí',
    'CE': 'Ceará',
    'RN': 'Rio Grande do Norte',
    'PB': 'Paraíba',
    'PE': 'Pernambuco',
    'AL': 'Alagoas',
    'SE': 'Sergipe',
    'BA': 'Bahia',
    'MG': 'Minas Gerais',
    'ES': 'Espírito Santo',
    'RJ': 'Rio de Janeiro',
    'SP': 'São Paulo',
    'PR': 'Paraná',
    'SC': 'Santa Catarina',
    'RS': 'Rio Grande do Sul',
    'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso',
    'GO': 'Goiás',
    'DF': 'Distrito Federal'
}

result['Estado'] = result['Sigla'].map(siglas_estados)
result['Óbitos'] = result['Óbitos'].astype(int)
result = result[['Estado', 'Sigla', 'Ano', 'Óbitos']]
result
```

# ***Novo dataframe (numero de casos)***

```python
result1 = df1.groupby(['NU_ANO','SG_UF_NOT'])['CLASSI_FIN'].sum().reset_index()

anos = df['NU_ANO'].unique()
estados = df['SG_UF_NOT'].unique()
for ano in anos:
  for estado in estados:
    if not ((result1['NU_ANO'] == ano) & (result1['SG_UF_NOT'] == estado)).any():
      result1 = result1.append({'NU_ANO': ano, 'SG_UF_NOT': estado, 'CLASSI_FIN': 0}, ignore_index=True)

result1 = result1.rename(columns={'NU_ANO': 'Ano'})
result1 = result1.rename(columns={'SG_UF_NOT': 'Sigla'})
result1 = result1.rename(columns={'CLASSI_FIN': 'N_CASOS'})
result1


siglas_estados = {
    'RO': 'Rondônia',
    'AC': 'Acre',
    'AM': 'Amazonas',
    'RR': 'Roraima',
    'PA': 'Pará',
    'AP': 'Amapá',
    'TO': 'Tocantins',
    'MA': 'Maranhão',
    'PI': 'Piauí',
    'CE': 'Ceará',
    'RN': 'Rio Grande do Norte',
    'PB': 'Paraíba',
    'PE': 'Pernambuco',
    'AL': 'Alagoas',
    'SE': 'Sergipe',
    'BA': 'Bahia',
    'MG': 'Minas Gerais',
    'ES': 'Espírito Santo',
    'RJ': 'Rio de Janeiro',
    'SP': 'São Paulo',
    'PR': 'Paraná',
    'SC': 'Santa Catarina',
    'RS': 'Rio Grande do Sul',
    'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso',
    'GO': 'Goiás',
    'DF': 'Distrito Federal'
}

result1['Estado'] = result1['Sigla'].map(siglas_estados)
result1['N_CASOS'] = result1['N_CASOS'].astype(int)
result1 = result1[['Estado', 'Sigla', 'Ano', 'N_CASOS']]
result1
```

```python
df_joined = pd.merge(result1[['Estado','Sigla', 'Ano', 'N_CASOS']], result[['Estado','Sigla', 'Ano', 'Óbitos']], on=['Estado','Ano', 'Sigla'])

df_joined.to_csv('arquivo.csv', index=False)
df_joined
```
